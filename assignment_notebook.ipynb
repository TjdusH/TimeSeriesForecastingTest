````python
# ÏãúÌÄÄÏä§ Îç∞Ïù¥ÌÑ∞ ÏÉùÏÑ± (Îî•Îü¨ÎãùÏö©)
sequence_length = 30

def create_sequences(X, y, seq_len=30):
    X_seq, y_seq = [], []
    for i in range(len(X) - seq_len):
        X_seq.append(X[i:i+seq_len])
        y_seq.append(y[i+seq_len])
    return np.array(X_seq), np.array(y_seq)

X_train_seq, y_train_seq = create_sequences(X_train_scaled, y_train.values, sequence_length)
X_val_seq, y_val_seq = create_sequences(X_val_scaled, y_val.values, sequence_length)
X_test_seq, y_test_seq = create_sequences(X_test_scaled, y_test.values, sequence_length)

print(f"ÏãúÌÄÄÏä§ Îç∞Ïù¥ÌÑ∞ shape: {X_train_seq.shape}")
print(f"  - ÏÉòÌîå Ïàò: {X_train_seq.shape[0]}")
print(f"  - ÏãúÌÄÄÏä§ Í∏∏Ïù¥: {X_train_seq.shape[1]}")
print(f"  - ÌäπÏÑ± Ïàò: {X_train_seq.shape[2]}")
````

## 2. ÏòàÏ†ú: LSTM Î™®Îç∏ + ÌôïÎ•† Í∏∞Î∞ò Ìè¨ÏßÄÏÖò Ï°∞Ï†à Ï†ÑÎûµ

Ïù¥ ÏòàÏ†úÎäî 2Í∞ÄÏßÄ ÌïµÏã¨ ÏöîÏÜåÎ°ú Íµ¨ÏÑ±Îê©ÎãàÎã§:

1. **ÏòàÏ∏° Î™®Îç∏**: Í∞ÄÍ≤© ÏÉÅÏäπ/ÌïòÎùΩÏùÑ ÏòàÏ∏°ÌïòÍ≥† ÌôïÎ•† Ï†úÍ≥µ
2. **Ìä∏Î†àÏù¥Îî© Ï†ÑÎûµ**: ÏòàÏ∏° ÌôïÎ•†Ïóê Îî∞Îùº Ìà¨Ïûê ÎπÑÏú® Ï°∞Ï†à

### Ï†ÑÎûµ ÏòàÏãú
- ÏÉÅÏäπ ÌôïÎ•† 90% ‚Üí ÏûêÎ≥∏Ïùò 90% Ìà¨Ïûê
- ÏÉÅÏäπ ÌôïÎ•† 60% ‚Üí ÏûêÎ≥∏Ïùò 60% Ìà¨Ïûê
- ÏÉÅÏäπ ÌôïÎ•† 40% ‚Üí Ìà¨ÏûêÌïòÏßÄ ÏïäÏùå (ÌòÑÍ∏à Î≥¥Ïú†)

````python
# ÏòàÏ†ú Î™®Îç∏: LSTM Í∏∞Î∞ò 2-class Î∂ÑÎ•òÍ∏∞
class ExampleLSTMModel(nn.Module):
    """
    ÏòàÏ†ú: LSTM Í∏∞Î∞ò Í∞ÄÍ≤© Î∞©Ìñ• ÏòàÏ∏° Î™®Îç∏
    Ï∂úÎ†•: ÏÉÅÏäπ ÌôïÎ•† (0~1)
    """
    def __init__(self, input_size, hidden_size=64, dropout=0.2):
        super(ExampleLSTMModel, self).__init__()
        
        self.lstm1 = nn.LSTM(input_size, hidden_size, batch_first=True, num_layers=1)
        self.dropout1 = nn.Dropout(dropout)
        self.bn1 = nn.BatchNorm1d(hidden_size)
        
        self.lstm2 = nn.LSTM(hidden_size, hidden_size//2, batch_first=True, num_layers=1)
        self.dropout2 = nn.Dropout(dropout)
        self.bn2 = nn.BatchNorm1d(hidden_size//2)
        
        self.fc1 = nn.Linear(hidden_size//2, 16)
        self.relu = nn.ReLU()
        self.dropout3 = nn.Dropout(dropout)
        
        self.fc2 = nn.Linear(16, 1)
        self.sigmoid = nn.Sigmoid()
        
    def forward(self, x):
        lstm_out, _ = self.lstm1(x)
        lstm_out = self.dropout1(lstm_out)
        lstm_out = lstm_out.permute(0, 2, 1)
        lstm_out = self.bn1(lstm_out)
        lstm_out = lstm_out.permute(0, 2, 1)
        
        lstm_out, _ = self.lstm2(lstm_out)
        lstm_out = self.dropout2(lstm_out[:, -1, :])
        lstm_out = self.bn2(lstm_out)
        
        out = self.fc1(lstm_out)
        out = self.relu(out)
        out = self.dropout3(out)
        
        out = self.fc2(out)
        out = self.sigmoid(out)
        
        return out

# ÏòàÏ†ú Î™®Îç∏ ÏÉùÏÑ±
example_model = ExampleLSTMModel(
    input_size=X_train_seq.shape[2],
    hidden_size=64,
    dropout=0.2
).to(device)

print("ÏòàÏ†ú Î™®Îç∏ Íµ¨Ï°∞:")
print(example_model)
print(f"\nTotal parameters: {sum(p.numel() for p in example_model.parameters()):,}")
````

````python
# ÌïôÏäµ Ìï®Ïàò (Ï†úÍ≥µÎê®)
def train_model(model, train_loader, val_loader, epochs=100, lr=0.001, patience=15):
    model = model.to(device)
    criterion = nn.BCELoss()
    optimizer = torch.optim.Adam(model.parameters(), lr=lr)
    
    history = {'train_loss': [], 'val_loss': [], 'train_acc': [], 'val_acc': []}
    best_val_loss = float('inf')
    patience_counter = 0
    best_model_state = None
    
    for epoch in range(epochs):
        model.train()
        train_loss = 0
        train_correct = 0
        train_total = 0
        
        for batch_X, batch_y in train_loader:
            batch_X, batch_y = batch_X.to(device), batch_y.to(device)
            optimizer.zero_grad()
            outputs = model(batch_X)
            loss = criterion(outputs, batch_y.unsqueeze(1))
            loss.backward()
            optimizer.step()
            
            train_loss += loss.item()
            predicted = (outputs > 0.5).float()
            train_total += batch_y.size(0)
            train_correct += (predicted.squeeze() == batch_y).sum().item()
        
        avg_train_loss = train_loss / len(train_loader)
        train_acc = train_correct / train_total
        
        model.eval()
        val_loss = 0
        val_correct = 0
        val_total = 0
        
        with torch.no_grad():
            for batch_X, batch_y in val_loader:
                batch_X, batch_y = batch_X.to(device), batch_y.to(device)
                outputs = model(batch_X)
                loss = criterion(outputs, batch_y.unsqueeze(1))
                
                val_loss += loss.item()
                predicted = (outputs > 0.5).float()
                val_total += batch_y.size(0)
                val_correct += (predicted.squeeze() == batch_y).sum().item()
        
        avg_val_loss = val_loss / len(val_loader)
        val_acc = val_correct / val_total
        
        history['train_loss'].append(avg_train_loss)
        history['val_loss'].append(avg_val_loss)
        history['train_acc'].append(train_acc)
        history['val_acc'].append(val_acc)
        
        if (epoch + 1) % 10 == 0:
            print(f'Epoch [{epoch+1}/{epochs}], Train Loss: {avg_train_loss:.4f}, Val Loss: {avg_val_loss:.4f}, Train Acc: {train_acc:.4f}, Val Acc: {val_acc:.4f}')
        
        if avg_val_loss < best_val_loss:
            best_val_loss = avg_val_loss
            patience_counter = 0
            best_model_state = model.state_dict().copy()
        else:
            patience_counter += 1
            
        if patience_counter >= patience:
            print(f'Early stopping at epoch {epoch+1}')
            break
    
    if best_model_state is not None:
        model.load_state_dict(best_model_state)
    
    return history

print("‚úÖ ÌïôÏäµ Ìï®Ïàò Ï§ÄÎπÑ ÏôÑÎ£å!")
````

````python
# ÏòàÏ†ú Î™®Îç∏ ÌïôÏäµ
print("ÏòàÏ†ú Î™®Îç∏ ÌïôÏäµ ÏãúÏûë...")
example_history = train_model(
    model=example_model,
    train_loader=train_loader,
    val_loader=val_loader,
    epochs=100,
    lr=0.001,
    patience=15
)
print("\n‚úÖ ÏòàÏ†ú Î™®Îç∏ ÌïôÏäµ ÏôÑÎ£å!")
````

ÏòàÏ∏° ÌôïÎ•†Ïóê Îî∞Îùº Ìà¨Ïûê ÎπÑÏú®ÏùÑ Ï°∞Ï†àÌïòÎäî Ï†ÑÎûµÏûÖÎãàÎã§.

````python
# ÏòàÏ∏° Ìï®Ïàò (ÌôïÎ•† Ìè¨Ìï®)
def predict_with_probability(model, data_loader):
    """
    Î™®Îç∏ ÏòàÏ∏° Î∞è ÌôïÎ•† Î∞òÌôò
    """
    model.eval()
    predictions_prob = []
    
    with torch.no_grad():
        for batch_X, _ in data_loader:
            batch_X = batch_X.to(device)
            outputs = model(batch_X)
            predictions_prob.append(outputs.cpu().numpy())
    
    predictions_prob = np.vstack(predictions_prob).flatten()
    predictions = (predictions_prob > 0.5).astype(int)
    
    return predictions_prob, predictions

# ÏòàÏ†ú Î™®Îç∏ ÏòàÏ∏°
example_prob, example_pred = predict_with_probability(example_model, test_loader)

print(f"ÏòàÏ∏° ÏôÑÎ£å!")
print(f"ÏòàÏ∏° Ïàò: {len(example_pred)}")
print(f"ÏÉÅÏäπ ÏòàÏ∏°: {np.sum(example_pred == 1)}Í∞ú")
print(f"ÌïòÎùΩ ÏòàÏ∏°: {np.sum(example_pred == 0)}Í∞ú")
print(f"\nÌèâÍ∑† ÏÉÅÏäπ ÌôïÎ•†: {example_prob.mean():.2%}")
````

````python
# ÌôïÎ•† Í∏∞Î∞ò Ìä∏Î†àÏù¥Îî© Ï†ÑÎûµ
def simulate_probability_trading(predictions_prob, actual_prices, dates, 
                                 initial_capital=10000, transaction_fee=0.001,
                                 threshold=0.5, position_scaling=True):
    """
    ÌôïÎ•† Í∏∞Î∞ò Ìä∏Î†àÏù¥Îî© ÏãúÎÆ¨Î†àÏù¥ÏÖò
    
    Parameters:
    -----------
    predictions_prob : array
        ÏÉÅÏäπ ÌôïÎ•† (0~1)
    actual_prices : array
        Ïã§Ï†ú Í∞ÄÍ≤©
    dates : array
        ÎÇ†Ïßú
    initial_capital : float
        Ï¥àÍ∏∞ ÏûêÎ≥∏
    transaction_fee : float
        Í±∞Îûò ÏàòÏàòÎ£å
    threshold : float
        Îß§Ïàò Í≤∞Ï†ï ÏûÑÍ≥ÑÍ∞í (Í∏∞Î≥∏ 0.5)
    position_scaling : bool
        True: ÌôïÎ•†Ïóê ÎπÑÎ°ÄÌïòÏó¨ Ìà¨Ïûê ÎπÑÏú® Ï°∞Ï†à
        False: ÏûÑÍ≥ÑÍ∞í Ï¥àÍ≥º Ïãú Ï†ÑÏï° Ìà¨Ïûê
    """
    cash = initial_capital
    btc_holdings = 0
    portfolio_values = []
    trade_log = []
    
    for i in range(len(predictions_prob)):
        current_price = actual_prices[i]
        prob = predictions_prob[i]
        
        portfolio_value = cash + btc_holdings * current_price
        portfolio_values.append(portfolio_value)
        
        # ÎßàÏßÄÎßâ ÎÇ† Ï†ÑÎüâ Îß§ÎèÑ
        if i == len(predictions_prob) - 1:
            if btc_holdings > 0:
                sell_value = btc_holdings * current_price * (1 - transaction_fee)
                trade_log.append({
                    'date': dates[i],
                    'action': 'SELL_ALL',
                    'price': current_price,
                    'prob': prob,
                    'amount': btc_holdings,
                    'value': btc_holdings * current_price,
                    'fee': btc_holdings * current_price * transaction_fee
                })
                cash += sell_value
                btc_holdings = 0
            continue
        
        # Ìà¨Ïûê ÎπÑÏú® Í≤∞Ï†ï
        if position_scaling:
            # ÌôïÎ•† Í∏∞Î∞ò Ìè¨ÏßÄÏÖò Ï°∞Ï†à
            if prob > threshold:
                # ÌôïÎ•†Ïóê ÎπÑÎ°ÄÌïòÏó¨ Ìà¨Ïûê (Ïòà: 70% ÌôïÎ•† ‚Üí 70% Ìà¨Ïûê)
                invest_ratio = prob
            else:
                # ÏûÑÍ≥ÑÍ∞í ÎØ∏ÎßåÏù¥Î©¥ Îß§ÎèÑ ÎòêÎäî Í¥ÄÎßù
                invest_ratio = 0
        else:
            # Îã®Ïàú Ï†ÑÎûµ: ÏûÑÍ≥ÑÍ∞í Ï¥àÍ≥º Ïãú Ï†ÑÏï° Ìà¨Ïûê
            invest_ratio = 1.0 if prob > threshold else 0.0
        
        # ÌòÑÏû¨ Ìè¨ÏßÄÏÖò ÎπÑÏú®
        current_position_ratio = (btc_holdings * current_price) / portfolio_value if portfolio_value > 0 else 0
        
        # Ìè¨ÏßÄÏÖò Ï°∞Ï†ï
        target_btc_value = portfolio_value * invest_ratio
        current_btc_value = btc_holdings * current_price
        
        if target_btc_value > current_btc_value:  # Îß§Ïàò ÌïÑÏöî
            buy_cash = target_btc_value - current_btc_value
            if buy_cash > cash:
                buy_cash = cash
            if buy_cash > 0:
                buy_amount = (buy_cash * (1 - transaction_fee)) / current_price
                btc_holdings += buy_amount
                trade_log.append({
                    'date': dates[i],
                    'action': 'BUY',
                    'price': current_price,
                    'prob': prob,
                    'amount': buy_amount,
                    'value': buy_cash,
                    'fee': buy_cash * transaction_fee
                })
                cash -= buy_cash
        elif target_btc_value < current_btc_value:  # Îß§ÎèÑ ÌïÑÏöî
            sell_btc = (current_btc_value - target_btc_value) / current_price
            if sell_btc > btc_holdings:
                sell_btc = btc_holdings
            if sell_btc > 0:
                sell_value = sell_btc * current_price * (1 - transaction_fee)
                trade_log.append({
                    'date': dates[i],
                    'action': 'SELL',
                    'price': current_price,
                    'prob': prob,
                    'amount': sell_btc,
                    'value': sell_btc * current_price,
                    'fee': sell_btc * current_price * transaction_fee
                })
                cash += sell_value
                btc_holdings -= sell_btc
    
    final_value = portfolio_values[-1]
    total_return = (final_value - initial_capital) / initial_capital * 100
    
    total_trade_volume = sum(trade['value'] for trade in trade_log)
    total_fees_paid = sum(trade['fee'] for trade in trade_log)
    
    return {
        'initial_capital': initial_capital,
        'final_value': final_value,
        'total_return': total_return,
        'portfolio_values': portfolio_values,
        'trade_log': trade_log,
        'num_trades': len(trade_log),
        'total_trade_volume': total_trade_volume,
        'total_fees_paid': total_fees_paid
    }

print("‚úÖ Ìä∏Î†àÏù¥Îî© ÏãúÎÆ¨Î†àÏù¥ÏÖò Ìï®Ïàò Ï§ÄÎπÑ ÏôÑÎ£å!")
````

```
   - ÎèôÏ†Å ÏûÑÍ≥ÑÍ∞í(Dynamic Threshold) Ï†ÑÎûµ: Î™®Îç∏Ïùò ÏòàÏ∏°Í∞íÏù¥ ÌäπÏ†ï Íµ¨Í∞ÑÏóê Ïè†Î¶¨Îäî ÌòÑÏÉÅÏùÑ Í∑πÎ≥µÌïòÍ∏∞ ÏúÑÌï¥, Í≥†Ï†ïÍ∞í(0.5) ÎåÄÏã† Îç∞Ïù¥ÌÑ∞ Î∂ÑÌè¨Ïùò 'ÏÉÅÏúÑ 20%'Î•º Îß§Ïàò, 'ÌïòÏúÑ 20%'Î•º Îß§ÎèÑ Í∏∞Ï§ÄÏúºÎ°ú ÏÑ§Ï†ï.
   - Ï°¥Î≤Ñ(Hysteresis): Îß§ÏàòÏôÄ Îß§ÎèÑ Í∏∞Ï§Ä ÏÇ¨Ïù¥Ïùò Íµ¨Í∞ÑÏóêÏÑúÎäî Îß§Îß§Î•º Ïâ¨Í≥† Ìè¨ÏßÄÏÖòÏùÑ Ïú†ÏßÄÌïòÏó¨ Ïû¶ÏùÄ Îß§Îß§Î°ú Ïù∏Ìïú ÏàòÏàòÎ£å ÏÜêÏã§ Î∞©ÏßÄ.

4. ÌïòÏù¥ÌçºÌååÎùºÎØ∏ÌÑ∞:
   - hidden_size: 128 (ÌëúÌòÑÎ†• Í∞ïÌôîÎ•º ÏúÑÌï¥ Í∏∞Ï°¥ 64ÏóêÏÑú 2Î∞∞ ÌôïÏû•)
   - learning_rate: 0.0005 (ÏïàÏ†ïÏ†ÅÏù∏ ÌïôÏäµÏùÑ ÏúÑÌï¥ ÎÇÆÍ≤å ÏÑ§Ï†ï)
   - threshold: Dynamic (Îß§Ïàò: ÏÉÅÏúÑ 20% / Îß§ÎèÑ: ÌïòÏúÑ 20%)
   - position_scaling: False (Ïã†Ìò∏ Î∞úÏÉù Ïãú Ï†ÑÏï° Îß§Ïàò/Îß§ÎèÑÌïòÎäî Ï∂îÏÑ∏ Ï∂îÏ¢Ö Î∞©Ïãù)

5. ÏòàÏ†úÏôÄÏùò Ï∞®Î≥ÑÏ†ê:
   - Î™®Îç∏: ÏòàÏ†ú(Îã®Ï∏µ LSTM)ÏôÄ Îã¨Î¶¨ 2Ï∏µ GRU Íµ¨Ï°∞Î•º ÏÇ¨Ïö©ÌïòÏó¨ ÍπäÏù¥ ÏûàÎäî Ìå®ÌÑ¥ ÌïôÏäµÏùÑ Ïú†ÎèÑ.
   - Ï†ÑÎûµ: ÏòàÏ†úÎäî Ï†àÎåÄÏ†ÅÏù∏ ÌôïÎ•†(0.5)Ïóê ÏùòÏ°¥ÌñàÏúºÎÇò, Î≥∏ Ï†ÑÎûµÏùÄ ÏòàÏ∏° ÌôïÎ•†Ïùò 'ÏÉÅÎåÄÏ†Å ÏàúÏúÑ(Percentile)'Î•º ÌôúÏö©ÌïòÏó¨ Î™®Îç∏Ïùò Ïò§Ï∞®Î•º Ï†ÑÎûµÏ†ÅÏúºÎ°ú Î≥¥Ï†ï.
```

````python
# ÏûêÏã†ÎßåÏùò Î™®Îç∏ Íµ¨ÌòÑ

class MyTradingModel(nn.Module):
    """
    Í∞úÏÑ†Îêú Î™®Îç∏: Stacked GRU + Dropout
    - LSTMÎ≥¥Îã§ ÌååÎùºÎØ∏ÌÑ∞Í∞Ä Ï†ÅÏñ¥ ÌïôÏäµÏù¥ Îπ†Î•¥Í≥† Í≥ºÏ†ÅÌï© ÏúÑÌóòÏù¥ Ï†ÅÏùå
    - 2Ï∏µ Íµ¨Ï°∞Î°ú Î≥µÏû°Ìïú Ìå®ÌÑ¥ ÌïôÏäµ Îä•Î†• Í∞ïÌôî
    """
    def __init__(self, input_size, hidden_size=128, dropout=0.):
        super(MyTradingModel, self).__init__()
        
        # LSTM ÎåÄÏã† GRU ÏÇ¨Ïö© (ÏÜçÎèÑ Î∞è ÏÑ±Îä• Í∞úÏÑ†)
        # num_layers=2: Ï∏µÏùÑ 2Í∞ú ÏåìÏïÑÏÑú Îçî ÍπäÏùÄ Ìå®ÌÑ¥ ÌïôÏäµ
        self.gru = nn.GRU(
            input_size, 
            hidden_size, 
            num_layers=2, 
            batch_first=True, 
            dropout=dropout
        )
        
        # Ï†ïÍ∑úÌôî Î∞è Í≥ºÏ†ÅÌï© Î∞©ÏßÄ Ï∏µ
        self.layer_norm = nn.LayerNorm(hidden_size)
        self.dropout = nn.Dropout(dropout)
        
        # Î∂ÑÎ•òÍ∏∞ (Classifier)
        self.fc1 = nn.Linear(hidden_size, 64)
        self.relu = nn.ReLU()
        self.fc2 = nn.Linear(64, 1)
        self.sigmoid = nn.Sigmoid()
        
    def forward(self, x):
        # x shape: (batch, seq_len, input_size)
        out, _ = self.gru(x)
        
        # ÎßàÏßÄÎßâ ÏãúÏ†ê(Time step)Ïùò ÌûàÎì† Ïä§ÌÖåÏù¥Ìä∏Îßå ÏÇ¨Ïö©
        out = out[:, -1, :] 
        
        out = self.layer_norm(out) # Ï∏µ Ï†ïÍ∑úÌôî
        out = self.dropout(out)
        
        out = self.fc1(out)
        out = self.relu(out)
        out = self.dropout(out)
        
        out = self.fc2(out)
        return self.sigmoid(out)

# Î™®Îç∏ ÏÉùÏÑ± (ÌïòÏù¥ÌçºÌååÎùºÎØ∏ÌÑ∞ ÌäúÎãùÎê®)
my_model = MyTradingModel(
    input_size=X_train_seq.shape[2],
    hidden_size=128,   # ÌûàÎì† ÏÇ¨Ïù¥Ï¶à Ï¶ùÍ∞Ä (ÌëúÌòÑÎ†• Í∞ïÌôî)
    dropout=0.3        # ÎìúÎ°≠ÏïÑÏõÉ Ï¶ùÍ∞Ä (ÏùºÎ∞òÌôî ÏÑ±Îä• Í∞ïÌôî)
).to(device)

print("ÎÇòÏùò Î™®Îç∏ Íµ¨Ï°∞:")
print(my_model)
print(f"\nTotal parameters: {sum(p.numel() for p in my_model.parameters()):,}")
````

````python
# Ìä∏Î†àÏù¥Îî© Ï†ÑÎûµ Ìï®Ïàò ÏàòÏ†ï

def simulate_dynamic_threshold_strategy(predictions_prob, actual_prices, dates, 
                                        initial_capital=10000, transaction_fee=0.001,
                                        buy_percentile=80, sell_percentile=20):
    """
    [ÎèôÏ†Å ÏûÑÍ≥ÑÍ∞í Ï†ÑÎûµ ÏãúÎÆ¨Î†àÏù¥ÏÖò]
    
    Í∏∞Ï°¥Ïùò Í≥†Ï†ïÎêú ÏûÑÍ≥ÑÍ∞í(0.5) ÎåÄÏã†, ÏòàÏ∏°Í∞íÎì§Ïùò Î∂ÑÌè¨Î•º Î∂ÑÏÑùÌïòÏó¨ 
    ÏÉÅÎåÄÏ†ÅÏúºÎ°ú ÎÜíÏùÄ Ï†êÏàòÎ•º 'Îß§Ïàò', ÎÇÆÏùÄ Ï†êÏàòÎ•º 'Îß§ÎèÑ' Ïã†Ìò∏Î°ú ÏÇ¨Ïö©ÌïòÎäî Ï†ÑÎûµ.
    
    Args:
        predictions_prob (list): Î™®Îç∏Ïù¥ ÏòàÏ∏°Ìïú ÏÉÅÏäπ ÌôïÎ•† Î¶¨Ïä§Ìä∏ (0~1 ÏÇ¨Ïù¥Í∞í)
        actual_prices (list): Ïã§Ï†ú ÎπÑÌä∏ÏΩîÏù∏ Í∞ÄÍ≤© Î¶¨Ïä§Ìä∏
        dates (list): ÎÇ†Ïßú Î¶¨Ïä§Ìä∏
        initial_capital (float): Ï¥àÍ∏∞ Ìà¨Ïûê ÏûêÎ≥∏Í∏à (Í∏∞Î≥∏ $10,000)
        transaction_fee (float): Í±∞Îûò ÏàòÏàòÎ£åÏú® (Í∏∞Î≥∏ 0.1%)
        buy_percentile (int): Îß§Ïàò Í∏∞Ï§Ä Î∞±Î∂ÑÏúÑÏàò (Í∏∞Î≥∏ 80 -> ÏÉÅÏúÑ 20%Ïùº Îïå Îß§Ïàò)
        sell_percentile (int): Îß§ÎèÑ Í∏∞Ï§Ä Î∞±Î∂ÑÏúÑÏàò (Í∏∞Î≥∏ 20 -> ÌïòÏúÑ 20%Ïùº Îïå Îß§ÎèÑ)
        
    Returns:
        dict: ÏµúÏ¢Ö ÏûêÏÇ∞, ÏàòÏùµÎ•†, Í±∞Îûò Í∏∞Î°ù Îì±Ïù¥ Îã¥Í∏¥ ÎîïÏÖîÎÑàÎ¶¨
    """
    
    # 1. ÎèôÏ†ÅÏúºÎ°ú ÏûÑÍ≥ÑÍ∞í Í≤∞Ï†ï (ÏÉÅÎåÄÌèâÍ∞Ä)
    # Î™®Îç∏Ïùò ÏòàÏ∏°Í∞í Ï†ÑÏ≤¥ Î∂ÑÌè¨ÏóêÏÑú ÏÉÅÏúÑ n%ÏôÄ ÌïòÏúÑ n%Ïóê Ìï¥ÎãπÌïòÎäî Ï†êÏàòÎ•º Ï∞æÎäîÎã§.
    # Ïòà: ÏòàÏ∏°Í∞íÏù¥ Ï†ÑÏ≤¥Ï†ÅÏúºÎ°ú 0.4~0.6 ÏÇ¨Ïù¥Ïóê Î™∞Î†§ÏûàÏñ¥ÎèÑ, 0.55Í∞Ä ÏÉÅÏúÑ 20%ÎùºÎ©¥ 0.55Î•º Îß§Ïàò Í∏∞Ï§ÄÏúºÎ°ú Ïû°ÎäîÎã§ (ÏÉÅÎåÄÌèâÍ∞Ä).
    buy_threshold = np.percentile(predictions_prob, buy_percentile)
    sell_threshold = np.percentile(predictions_prob, sell_percentile)
    
    # ÏïàÏ†ÑÏû•Ïπò: ÏûÑÍ≥ÑÍ∞íÏù¥ Í≤πÏπòÎ©¥ Í±∞ÎûòÍ∞Ä Î∂àÍ∞ÄÎä•ÌïòÎØÄÎ°ú Í∞ïÏ†úÎ°ú Í∏∞Î≥∏Í∞íÏùÑ Ìï†ÎãπÌïòÏó¨ ÏóêÎü¨Î•º Î∞©ÏßÄ.
    if buy_threshold == sell_threshold:
        buy_threshold = 0.51
        sell_threshold = 0.49
        print("Í≤ΩÍ≥†: Î™®Îç∏ ÏòàÏ∏°Í∞íÏù¥ Î™®Îëê ÎèôÏùºÌï©ÎãàÎã§. ÌïôÏäµÏù¥ Ï†úÎåÄÎ°ú Ïïà Îêú Í≤É Í∞ôÏäµÎãàÎã§.")

    print(f"ÎèôÏ†Å ÏûÑÍ≥ÑÍ∞í Ï†ÅÏö©: Îß§Ïàò Í∏∞Ï§Ä {buy_threshold:.4f} Ïù¥ÏÉÅ / Îß§ÎèÑ Í∏∞Ï§Ä {sell_threshold:.4f} Ïù¥Ìïò")
    
    # 2. Ìè¨Ìä∏Ìè¥Î¶¨Ïò§ Ï¥àÍ∏∞Ìôî

    cash = initial_capital       # ÌòÑÏû¨ Î≥¥Ïú† ÌòÑÍ∏à
    btc_holdings = 0             # ÌòÑÏû¨ Î≥¥Ïú† ÎπÑÌä∏ÏΩîÏù∏ Í∞úÏàò
    portfolio_values = []        # Îß§ÏùºÎß§ÏùºÏùò Ï¥ù ÏûêÏÇ∞ Í∞ÄÏπò Í∏∞Î°ùÏö©
    trade_log = []               # Îß§Îß§ ÏùºÏßÄ Í∏∞Î°ùÏö©
    total_fees_paid = 0          # ÏàòÏàòÎ£å Ï¥ùÏï° (ÏÑ±Í≥º Î∂ÑÏÑùÏö©)
    
    # ÌòÑÏû¨ Ìè¨ÏßÄÏÖò ÏÉÅÌÉú (False: ÌòÑÍ∏à Î≥¥Ïú† Ï§ë / True: ÎπÑÌä∏ÏΩîÏù∏ Î≥¥Ïú† Ï§ë)
    has_position = False 
    
    # 3. ÏùºÎ≥Ñ ÏãúÎÆ¨Î†àÏù¥ÏÖò (Backtesting Loop)
    for i in range(len(predictions_prob)):
        current_price = actual_prices[i]  # Ïò§ÎäòÏùò Í∞ÄÍ≤©
        prob = predictions_prob[i]  # Ïò§ÎäòÏùò ÏòàÏ∏° ÌôïÎ•†
        
        # 3-1. Îß§Ïàò (ÏÉÅÏúÑ n% ÏïàÏóê Îì§Î©¥ Îß§Ïàò)
        if not has_position and prob >= buy_threshold:
            fee = cash * transaction_fee

            # Ïã§Ï†ú Îß§Ïàò Í∞ÄÎä• Í∏àÏï°ÏúºÎ°ú ÎπÑÌä∏ÏΩîÏù∏ Îß§Ïàò (ÏàòÏàòÎ£å Ï†úÏô∏ ÌõÑ)
            buy_amount = (cash - fee) / current_price
            
            # ÏÉÅÌÉú ÏóÖÎç∞Ïù¥Ìä∏
            btc_holdings = buy_amount
            cash = 0
            has_position = True
            total_fees_paid += fee
            
            trade_log.append({
                'date': dates[i], 'action': 'BUY', 'price': current_price, 
                'prob': prob, 'value': btc_holdings * current_price, 'fee': fee
            })

        # 3-2. Îß§ÎèÑ (ÌïòÏúÑ n% ÏïàÏóê Îì§Î©¥ Îß§ÎèÑ)
        elif has_position and prob <= sell_threshold:
            # ÌòÑÏû¨ Î≥¥Ïú†Ìïú ÎπÑÌä∏ÏΩîÏù∏Ïùò Ï¥ù Í∞ÄÏπò Í≥ÑÏÇ∞
            gross_value = btc_holdings * current_price

            fee = gross_value * transaction_fee

            # ÏàòÏàòÎ£åÎ•º Î∫Ä Ïã§Ï†ú ÌòÑÍ∏à ÏàòÎ†πÏï°
            sell_value = gross_value - fee
            
            # ÏÉÅÌÉú ÏóÖÎç∞Ïù¥Ìä∏
            cash = sell_value
            btc_holdings = 0
            has_position = False
            total_fees_paid += fee
            
            trade_log.append({
                'date': dates[i], 'action': 'SELL', 'price': current_price, 
                'prob': prob, 'value': sell_value, 'fee': fee
            })
            
        # 3. Î≤ÑÌã∞Í∏∞
        # ÏòàÏ∏° ÌôïÎ•†Ïù¥ Ïï†Îß§Ìïú Íµ¨Í∞Ñ (Îß§ÎèÑ Í∏∞Ï§Ä < ÌôïÎ•† < Îß§Ïàò Í∏∞Ï§Ä)
        # -> ÏïÑÎ¨¥ ÌñâÎèôÎèÑ ÌïòÏßÄ ÏïäÍ≥† ÌòÑÏû¨ ÏÉÅÌÉú(ÌòÑÍ∏à or ÏΩîÏù∏)Î•º Ïú†ÏßÄÌï® (Ï∂îÏÑ∏ Ï∂îÏ¢Ö Ìö®Í≥º)
        
        if has_position:
            current_val = btc_holdings * current_price
        else:
            current_val = cash
        portfolio_values.append(current_val)

    if has_position:
        final_val = btc_holdings * actual_prices[-1]
        portfolio_values[-1] = final_val

    total_return = (portfolio_values[-1] - initial_capital) / initial_capital * 100
    
    return {
        'initial_capital': initial_capital,
        'final_value': portfolio_values[-1],
        'total_return': total_return,
        'portfolio_values': portfolio_values,
        'num_trades': len(trade_log),
        'trade_log': trade_log,
        'total_fees_paid': total_fees_paid
    }
````

````python
# ÏòàÏ∏° Î∞è Ìä∏Î†àÏù¥Îî© ÏãúÎÆ¨Î†àÏù¥ÏÖò

# ÏòàÏ∏°
my_prob, my_pred = predict_with_probability(my_model, test_loader)

# Ìä∏Î†àÏù¥Îî© ÏãúÎÆ¨Î†àÏù¥ÏÖò
my_result = simulate_dynamic_threshold_strategy(  #ÏàòÏ†ïÌïú Ï†ÑÎûµ Ïã§Ìñâ
    predictions_prob=my_prob,
    actual_prices=test_prices_aligned,
    dates=test_dates_aligned,
    initial_capital=10000,
    transaction_fee=0.001,
    buy_percentile=80,  # ÏÉÅÏúÑ 20%
    sell_percentile=20  # ÌïòÏúÑ 20%
)

print("="*70)
print("ÎÇòÏùò Ìä∏Î†àÏù¥Îî© Ï†ÑÎûµ Í≤∞Í≥º")
print("="*70)
print(f"Ï¥àÍ∏∞ ÏûêÎ≥∏: ${my_result['initial_capital']:,.2f}")
print(f"ÏµúÏ¢Ö ÏûêÎ≥∏: ${my_result['final_value']:,.2f}")
print(f"ÏàòÏùµÎ•†: {my_result['total_return']:.2f}%")
print(f"Buy and Hold ÎåÄÎπÑ: {my_result['total_return'] - buy_hold_return:.2f}%p")
print(f"Í±∞Îûò ÌöüÏàò: {my_result['num_trades']}Ìöå")
print(f"Ï¥ù ÏàòÏàòÎ£å: ${my_result['total_fees_paid']:,.2f}")
print("="*70)
````

